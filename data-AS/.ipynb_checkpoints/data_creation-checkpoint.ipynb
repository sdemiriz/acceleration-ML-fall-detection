```python
mean_laying_X=np.zeros(20)
mean_laying_Y=np.zeros(20)
mean_laying_Z=np.zeros(20)
for j in range (20):
    mean_laying_X[j]= files_laying[j]['ax (m/s^2)'].mean()
    mean_laying_Y[j]= files_laying[j]['ay (m/s^2)'].mean()
    mean_laying_Z[j]= files_laying[j]['az (m/s^2)'].mean()
    
mean_phonefall_X=np.zeros(20)
mean_phonefall_Y=np.zeros(20)
mean_phonefall_Z=np.zeros(20)
for j in range (20):
    mean_phonefall_X[j]= files_phonefall[j]['ax (m/s^2)'].mean()
    mean_phonefall_Y[j]= files_phonefall[j]['ay (m/s^2)'].mean()
    mean_phonefall_Z[j]= files_phonefall[j]['az (m/s^2)'].mean()
    
mean_sit_X=np.zeros(20)
mean_sit_Y=np.zeros(20)
mean_sit_Z=np.zeros(20)
for j in range (20):
    mean_sit_X[j]= files_sit[j]['ax (m/s^2)'].mean()
    mean_sit_Y[j]= files_sit[j]['ay (m/s^2)'].mean()
    mean_sit_Z[j]= files_sit[j]['az (m/s^2)'].mean()
    
mean_walkfall_X=np.zeros(20)
mean_walkfall_Y=np.zeros(20)
mean_walkfall_Z=np.zeros(20)
for j in range (20):
    mean_walkfall_X[j]= files_walkfall[j]['ax (m/s^2)'].mean()
    mean_walkfall_Y[j]= files_walkfall[j]['ay (m/s^2)'].mean()
    mean_walkfall_Z[j]= files_walkfall[j]['az (m/s^2)'].mean() 
```

```python
store_X= files_laying[0]['ax (m/s^2)']
for i in range(19):
    store_X= store_X.append(files_laying[i+1]['ax (m/s^2)'])
    
store_Y= files_laying[1]['ay (m/s^2)']
for i in range(19):
    store_Y= store_Y.append(files_laying[i+1]['ay (m/s^2)'])
    
store_Z= files_laying[2]['az (m/s^2)']
for i in range(19):
    store_Z= store_Z.append(files_laying[i+1]['az (m/s^2)'])
    

    
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
b, a = signal.butter(3, 0.1, btype='lowpass', analog=False)
low_passed_X = signal.filtfilt(b, a,store_X)
plt.ylabel('Range')
plt.xlabel('X')
plt.title('lay')
plt.plot(store_X)
plt.subplot(1, 2, 2)
plt.plot(low_passed_X)
plt.ylabel('Range')
plt.xlabel('X')
plt.title('lay')
plt.savefig('lay_X_filtered.png')

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
b, a = signal.butter(3, 0.1, btype='lowpass', analog=False)
plt.ylabel('Range')
plt.title('lay')
plt.xlabel('Y')
low_passed_Y = signal.filtfilt(b, a,store_Y)
plt.plot(store_Y)
plt.subplot(1, 2, 2)
plt.plot(low_passed_Y)
plt.ylabel('Range')
plt.title('lay')
plt.xlabel('Y')
plt.savefig('lay_Y_filtered.png')

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
b, a = signal.butter(3, 0.1, btype='lowpass', analog=False)
low_passed_Z = signal.filtfilt(b, a,store_Z)
plt.ylabel('Range')
plt.title('lay')
plt.xlabel('Z')
plt.plot(store_Z)
plt.subplot(1, 2, 2)
plt.plot(low_passed_Z)
plt.ylabel('Range')
plt.title('lay')
plt.xlabel('Z')
plt.savefig('lay_Z_filtered_lay.png')
```
