```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
path = r'/home/amritpal/sfuhome/falldetection353/fallDetection353/data3/laying' # path
all_files = glob.glob(path + "/*.csv")

files_laying = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    files_laying.append(df)

frame = pd.concat(files_laying, axis=0, ignore_index=True)
```
